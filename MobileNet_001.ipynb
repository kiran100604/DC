```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import torch
import torch.nn as nn
import torch.nn.functional as F
import torchvision
from torchvision import datasets, transforms, models
from torch.utils.data import Dataset, DataLoader
from torch.utils.data.sampler import SubsetRandomSampler
from torch.optim import lr_scheduler
from sklearn.metrics import cohen_kappa_score, accuracy_score
from PIL import Image
import cv2
from tqdm import tqdm
import copy

# Check if GPU is available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using {device} device")

BASE_DIR = "/content/"
TRAIN_CSV = os.path.join(BASE_DIR, "/kaggle/input/aptos2019-blindness-detection/train.csv")
TEST_CSV = os.path.join(BASE_DIR, "/kaggle/input/aptos2019-blindness-detection/train.csv")
TRAIN_IMGS = os.path.join(BASE_DIR, "/kaggle/input/aptos2019-blindness-detection/train_images")
TEST_IMGS = os.path.join(BASE_DIR, "/kaggle/input/aptos2019-blindness-detection/test_images")

# Create folders if they don't exist
os.makedirs(BASE_DIR, exist_ok=True)
os.makedirs(TRAIN_IMGS, exist_ok=True)
os.makedirs(TEST_IMGS, exist_ok=True)

# Load the data
train_df = pd.read_csv(TRAIN_CSV)
test_df = pd.read_csv(TEST_CSV)

print(f'Train Size = {len(train_df)}')
print(f'Test Size = {len(test_df)}')

# Display data distribution
def plot_class_distribution(df):
    counts = df['diagnosis'].value_counts()
    class_list = ['No DR', 'Mild', 'Moderate', 'Severe', 'Proliferative DR']

    # Create a mapping for visualization
    mapping = {}
    for i, label in enumerate(class_list):
        if i in counts.index:
            mapping[label] = counts[i]
        else:
            mapping[label] = 0

    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(mapping.keys()), y=list(mapping.values()), alpha=0.8, palette='bright')
    plt.title('Distribution of Output Classes')
    plt.ylabel('Number of Occurrences', fontsize=12)
    plt.xlabel('Target Classes', fontsize=12)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


# plot_class_distribution(train_df)

# Image preprocessing and augmentation
class EyeDataset(Dataset):
    def __init__(self, df, img_dir, transform=None, test=False):
        self.df = df
        self.img_dir = img_dir
        self.transform = transform
        self.test = test

    def __len__(self):
        return len(self.df)

    def __getitem__(self, idx):
        img_name = self.df.iloc[idx]['id_code']
        img_path = os.path.join(self.img_dir, f"{img_name}.png")

        # Read image using OpenCV for preprocessing
        image = cv2.imread(img_path)

        # Image preprocessing (crop black borders, enhance contrast)
        image = self.preprocess_image(image)

        if self.transform:
            image = self.transform(image)

        if self.test:
            return image
        else:
            label = self.df.iloc[idx]['diagnosis']
            return image, label

    def preprocess_image(self, image):
        # Convert to RGB (OpenCV uses BGR)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # Convert to grayscale for preprocessing
        gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

        # Find the non-black region (crop black borders)
        mask = gray > 10
        if mask.any():
            rows = np.any(mask, axis=1)
            cols = np.any(mask, axis=0)
            rmin, rmax = np.where(rows)[0][[0, -1]]
            cmin, cmax = np.where(cols)[0][[0, -1]]

            # Add some padding
            padding = 10
            rmin = max(0, rmin - padding)
            rmax = min(image.shape[0], rmax + padding)
            cmin = max(0, cmin - padding)
            cmax = min(image.shape[1], cmax + padding)

            # Crop the image
            image = image[rmin:rmax, cmin:cmax]

        # Apply CLAHE for contrast enhancement
        lab = cv2.cvtColor(image, cv2.COLOR_RGB2LAB)
        l, a, b = cv2.split(lab)
        clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
        l = clahe.apply(l)
        lab = cv2.merge((l, a, b))
        image = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)

        return image

# Define transforms with more augmentations for regularization
train_transforms = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.5),
    transforms.RandomRotation(20),
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.05),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

valid_transforms = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Create datasets
train_dataset = EyeDataset(train_df, TRAIN_IMGS, transform=train_transforms)

# Split into train and validation sets
valid_size = 0.2
num_train = len(train_dataset)
indices = list(range(num_train))
np.random.seed(42)  # For reproducibility
np.random.shuffle(indices)
split = int(np.floor(valid_size * num_train))
train_idx, valid_idx = indices[split:], indices[:split]

train_sampler = SubsetRandomSampler(train_idx)
valid_sampler = SubsetRandomSampler(valid_idx)

# Create data loaders
batch_size = 32  # Smaller batch size for better generalization
train_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=train_sampler, num_workers=2)
valid_loader = DataLoader(train_dataset, batch_size=batch_size, sampler=valid_sampler, num_workers=2)

print(f"Number of training batches: {len(train_loader)}")
print(f"Number of validation batches: {len(valid_loader)}")

# Define Teacher model (EfficientNet)
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights

class EfficientDRModel(nn.Module):
    def __init__(self, num_classes=5):
        super(EfficientDRModel, self).__init__()
        # Use pre-trained EfficientNet-B0
        self.effnet = efficientnet_b0(weights=EfficientNet_B0_Weights.IMAGENET1K_V1)

        # Replace classifier with custom head
        num_ftrs = self.effnet.classifier[1].in_features
        self.effnet.classifier = nn.Identity()

        # Add custom classifier with dropout for regularization
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(num_ftrs, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        features = self.effnet(x)
        return self.classifier(features)

# Define Student model (MobileNetV2)
from torchvision.models import mobilenet_v2, MobileNet_V2_Weights

class MobileNetDRModel(nn.Module):
    def __init__(self, num_classes=5):
        super(MobileNetDRModel, self).__init__()
        # Use pre-trained MobileNetV2
        self.mobilenet = mobilenet_v2(weights=MobileNet_V2_Weights.IMAGENET1K_V1)

        # Replace classifier with custom head
        num_ftrs = self.mobilenet.classifier[1].in_features
        self.mobilenet.classifier = nn.Identity()

        # Add custom classifier with dropout for regularization
        self.classifier = nn.Sequential(
            nn.Dropout(0.3),
            nn.Linear(num_ftrs, 256),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.Linear(256, num_classes)
        )

    def forward(self, x):
        features = self.mobilenet(x)
        return self.classifier(features)

# Initialize teacher model and load weights if available
teacher_model = EfficientDRModel(num_classes=5).to(device)

# Try to load the trained EfficientNet model if available
try:
    checkpoint = torch.load('/kaggle/input/h/pytorch/default/1/best_dr_model.pt')
    teacher_model.load_state_dict(checkpoint['model_state_dict'])
    print("Loaded teacher model weights successfully!")
    teacher_kappa = checkpoint.get('kappa', 0.0)
    print(f"Teacher model kappa: {teacher_kappa:.4f}")
except FileNotFoundError:
    print("Teacher model weights not found. Please train the teacher model first.")
    # If you want to proceed without a pre-trained teacher, remove the next line
    # raise

# Initialize student model
student_model = MobileNetDRModel(num_classes=5).to(device)

# Calculate parameters for comparison
def count_parameters(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

teacher_params = count_parameters(teacher_model)
student_params = count_parameters(student_model)
print(f"Teacher model parameters: {teacher_params:,}")
print(f"Student model parameters: {student_params:,}")
print(f"Parameter reduction: {(1 - student_params/teacher_params)*100:.2f}%")

# Knowledge Distillation Loss Function
class DistillationLoss(nn.Module):
    def __init__(self, alpha=0.5, temperature=3.0):
        super(DistillationLoss, self).__init__()
        self.alpha = alpha  # Weight for distillation loss
        self.temperature = temperature  # Temperature for softening probabilities
        self.criterion = nn.CrossEntropyLoss(label_smoothing=0.1)  # Hard target loss

    def forward(self, outputs, labels, teacher_outputs):
        # Hard target loss
        hard_loss = self.criterion(outputs, labels)

        # Soft target loss (KL divergence)
        soft_loss = nn.KLDivLoss(reduction='batchmean')(
            F.log_softmax(outputs / self.temperature, dim=1),
            F.softmax(teacher_outputs / self.temperature, dim=1)
        ) * (self.temperature ** 2)

        # Combined loss
        loss = (1 - self.alpha) * hard_loss + self.alpha * soft_loss
        return loss

# Function to calculate quadratic weighted kappa score
def quadratic_kappa(y_true, y_pred):
    return cohen_kappa_score(y_true, y_pred, weights='quadratic')

# Use AdamW optimizer with weight decay for regularization
optimizer = torch.optim.AdamW(student_model.parameters(), lr=1e-4, weight_decay=1e-5)

# OneCycleLR scheduler for faster convergence
scheduler = torch.optim.lr_scheduler.OneCycleLR(
    optimizer,
    max_lr=1e-3,
    epochs=15,  # Increased epochs for distillation
    steps_per_epoch=len(train_loader),
    pct_start=0.2
)

# Mixed precision training for faster computation on GPU
scaler = torch.cuda.amp.GradScaler() if device.type == 'cuda' else None

# Define distillation criterion
# distillation_criterion = DistillationLoss(alpha=0.5, temperature=3.0)

# Training function with knowledge distillation
def train_student_model(student_model, teacher_model, criterion, optimizer, scheduler, num_epochs=20, patience=5):
    best_kappa = 0.0
    best_model_wts = None
    no_improve_epochs = 0

    # History for plotting
    history = {
        'train_loss': [], 'val_loss': [],
        'train_acc': [], 'val_acc': [],
        'val_kappa': []
    }

    # Set teacher model to eval mode
    teacher_model.eval()

    for epoch in range(num_epochs):
        print(f'Epoch {epoch+1}/{num_epochs}')
        print('-' * 10)

        # Each epoch has a training and validation phase
        for phase in ['train', 'val']:
            if phase == 'train':
                student_model.train()
                dataloader = train_loader
            else:
                student_model.eval()
                dataloader = valid_loader

            running_loss = 0.0
            all_labels = []
            all_preds = []

            # Iterate over data
            for inputs, labels in tqdm(dataloader, desc=phase):
                inputs = inputs.to(device)
                labels = labels.to(device)

                # Zero the parameter gradients
                optimizer.zero_grad()

                # Forward
                with torch.set_grad_enabled(phase == 'train'):
                    with torch.cuda.amp.autocast() if scaler else torch.no_grad():
                        # Student predictions
                        student_outputs = student_model(inputs)

                        # Get teacher predictions - no grad needed
                        with torch.no_grad():
                            teacher_outputs = teacher_model(inputs)

                        # Calculate loss with knowledge distillation
                        if phase == 'train':
                            loss = criterion(student_outputs, labels, teacher_outputs)
                        else:
                            # Just use standard cross-entropy for validation
                            loss = nn.CrossEntropyLoss()(student_outputs, labels)

                    # Backward + optimize only if in training phase
                    if phase == 'train':
                        if scaler:
                            scaler.scale(loss).backward()
                            scaler.step(optimizer)
                            scaler.update()
                        else:
                            loss.backward()
                            optimizer.step()
                        scheduler.step()

                # Statistics
                running_loss += loss.item() * inputs.size(0)
                _, preds = torch.max(student_outputs, 1)
                all_labels.extend(labels.cpu().numpy())
                all_preds.extend(preds.cpu().numpy())

            epoch_loss = running_loss / len(dataloader.sampler)
            epoch_acc = accuracy_score(all_labels, all_preds)

            # Calculate quadratic kappa for validation
            if phase == 'val':
                epoch_kappa = quadratic_kappa(all_labels, all_preds)
                print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f} Kappa: {epoch_kappa:.4f}')
                history['val_loss'].append(epoch_loss)
                history['val_acc'].append(epoch_acc)
                history['val_kappa'].append(epoch_kappa)

                # Deep copy the model if best validation kappa so far
                if epoch_kappa > best_kappa:
                    best_kappa = epoch_kappa
                    best_model_wts = copy.deepcopy(student_model.state_dict())
                    no_improve_epochs = 0

                    # Save the best model
                    torch.save({
                        'epoch': epoch,
                        'model_state_dict': student_model.state_dict(),
                        'optimizer_state_dict': optimizer.state_dict(),
                        'kappa': best_kappa
                    }, 'best_mobilenet_dr_model.pt')
                    print(f'Saved MobileNet model with improved kappa: {best_kappa:.4f}')
                else:
                    no_improve_epochs += 1
            else:
                print(f'{phase} Loss: {epoch_loss:.4f} Acc: {epoch_acc:.4f}')
                history['train_loss'].append(epoch_loss)
                history['train_acc'].append(epoch_acc)

        print()

        # Early stopping
        if no_improve_epochs >= patience:
            print(f'Early stopping after {epoch+1} epochs with no improvement')
            break

    # Load best model weights
    student_model.load_state_dict(best_model_wts)
    return student_model, history

# Train the student model with knowledge distillation
# student_model, history = train_student_model(
#     student_model,
#     teacher_model,
#     distillation_criterion,
#     optimizer,
#     scheduler,
#     num_epochs=20,
#     patience=5
# )

# Define distillation criterion with alpha=0.1
# distillation_criterion_01 = DistillationLoss(alpha=0.1, temperature=3.0)

# # Train with alpha=0.1
# student_model_01, history_01 = train_student_model(
#     student_model,
#     teacher_model,
#     distillation_criterion_01,  # Using 0.1 weight
#     optimizer,
#     scheduler,
#     num_epochs=20,
#     patience=5
# )

# Define distillation criterion with alpha=0.01
distillation_criterion_001 = DistillationLoss(alpha=0.01, temperature=3.0)

# Reset or create a new student model
student_model_001 = MobileNetDRModel(num_classes=5).to(device)
optimizer_001 = torch.optim.AdamW(student_model_001.parameters(), lr=1e-4, weight_decay=1e-5)
scheduler_001 = torch.optim.lr_scheduler.OneCycleLR(
    optimizer_001,
    max_lr=1e-3,
    epochs=20,
    steps_per_epoch=len(train_loader),
    pct_start=0.2
)

# Train with alpha=0.01
student_model_001, history_001 = train_student_model(
    student_model_001,
    teacher_model,
    distillation_criterion_001,  # Using 0.01 weight
    optimizer_001,
    scheduler_001,
    num_epochs=20,
    patience=5
)

# Plot training history
# plt.figure(figsize=(15, 5))

# plt.subplot(1, 3, 1)
# plt.plot(history['train_loss'], label='Train Loss')
# plt.plot(history['val_loss'], label='Validation Loss')
# plt.title('MobileNet Loss Curves')
# plt.xlabel('Epoch')
# plt.ylabel('Loss')
# plt.legend()

# plt.subplot(1, 3, 2)
# plt.plot(history['train_acc'], label='Train Accuracy')
# plt.plot(history['val_acc'], label='Validation Accuracy')
# plt.title('MobileNet Accuracy Curves')
# plt.xlabel('Epoch')
# plt.ylabel('Accuracy')
# plt.legend()

# plt.subplot(1, 3, 3)
# plt.plot(history['val_kappa'], label='Validation Kappa')
# plt.title('MobileNet Quadratic Weighted Kappa')
# plt.xlabel('Epoch')
# plt.ylabel('Kappa Score')
# plt.legend()

# plt.tight_layout()
# plt.show()

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.plot(history_001['train_loss'], label='Train Loss')
plt.plot(history_001['val_loss'], label='Validation Loss')
plt.title('MobileNet Loss Curves (Alpha=0.1)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 3, 2)
plt.plot(history_001['train_acc'], label='Train Accuracy')
plt.plot(history_001['val_acc'], label='Validation Accuracy')
plt.title('MobileNet Accuracy Curves (Alpha=0.1)')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 3, 3)
plt.plot(history_001['val_kappa'], label='Validation Kappa')
plt.title('MobileNet Quadratic Weighted Kappa (Alpha=0.1)')
plt.xlabel('Epoch')
plt.ylabel('Kappa Score')
plt.legend()

plt.tight_layout()
plt.show()

# Compare teacher and student models
def compare_models(teacher_model, student_model, valid_loader):
    """Compare performance between teacher and student models"""
    # Set both models to eval mode
    teacher_model.eval()
    student_model.eval()

    teacher_preds = []
    student_preds = []
    all_labels = []

    with torch.no_grad():
        for inputs, labels in tqdm(valid_loader, desc='Comparing models'):
            inputs = inputs.to(device)
            labels = labels.to(device)

            # Teacher predictions
            teacher_outputs = teacher_model(inputs)
            _, teacher_pred = torch.max(teacher_outputs, 1)

            # Student predictions
            student_outputs = student_model(inputs)
            _, student_pred = torch.max(student_outputs, 1)

            # Store predictions and labels
            teacher_preds.extend(teacher_pred.cpu().numpy())
            student_preds.extend(student_pred.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    # Calculate metrics
    teacher_acc = accuracy_score(all_labels, teacher_preds)
    student_acc = accuracy_score(all_labels, student_preds)

    teacher_kappa = quadratic_kappa(all_labels, teacher_preds)
    student_kappa = quadratic_kappa(all_labels, student_preds)

    agreement = accuracy_score(teacher_preds, student_preds)

    print("Model Comparison:")
    print(f"Teacher Accuracy: {teacher_acc:.4f}, Kappa: {teacher_kappa:.4f}")
    print(f"Student Accuracy: {student_acc:.4f}, Kappa: {student_kappa:.4f}")
    print(f"Agreement between Teacher and Student: {agreement:.4f}")

    return teacher_acc, student_acc, teacher_kappa, student_kappa, agreement

# Compare the models
teacher_acc, student_acc, teacher_kappa, student_kappa, agreement = compare_models(
    teacher_model, student_model_001, valid_loader
)

# Make predictions on the test dataset
test_dataset = EyeDataset(test_df, TEST_IMGS, transform=valid_transforms, test=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

def predict(model, test_loader):
    model.eval()
    all_preds = []

    with torch.no_grad():
        for inputs in tqdm(test_loader, desc='Predicting'):
            inputs = inputs.to(device)
            outputs = model(inputs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())

    return all_preds

# Make predictions with the student model
# predictions = predict(student_model, test_loader)

# Create submission file
# submission = pd.DataFrame({
#     'id_code': test_df['id_code'],
#     'diagnosis': predictions
# })
# submission.to_csv('mobilenet_submission.csv', index=False)
# print("MobileNet submission file created!")

# Model size comparison
def get_model_size(model):
    """Calculate model size in MB"""
    torch.save(model.state_dict(), "temp_model.pt")
    size_mb = os.path.getsize("temp_model.pt") / (1024 * 1024)
    os.remove("temp_model.pt")
    return size_mb

teacher_size = get_model_size(teacher_model)
student_size = get_model_size(student_model)

print(f"Teacher model size: {teacher_size:.2f} MB")
print(f"Student model size: {student_size:.2f} MB")
print(f"Size reduction: {(1 - student_size/teacher_size)*100:.2f}%")

# Inference time comparison
def measure_inference_time(model, input_size=(1, 3, 224, 224), num_iterations=100):
    """Measure inference time for a model"""
    model.eval()
    dummy_input = torch.randn(input_size).to(device)

    # Warm-up
    for _ in range(10):
        _ = model(dummy_input)

    # Measure
    start_time = torch.cuda.Event(enable_timing=True)
    end_time = torch.cuda.Event(enable_timing=True)

    start_time.record()
    for _ in range(num_iterations):
        _ = model(dummy_input)
    end_time.record()

    # Synchronize
    torch.cuda.synchronize()

    # Calculate average time
    avg_time = start_time.elapsed_time(end_time) / num_iterations
    return avg_time

teacher_time = measure_inference_time(teacher_model)
student_time = measure_inference_time(student_model)

print(f"Teacher inference time: {teacher_time:.2f} ms")
print(f"Student inference time: {student_time:.2f} ms")
print(f"Speed improvement: {(teacher_time/student_time):.2f}x")

# Plot comparison between teacher and student
metrics = ['Accuracy', 'Kappa', 'Size (MB)', 'Inference Time (ms)']
teacher_values = [teacher_acc, teacher_kappa, teacher_size, teacher_time]
student_values = [student_acc, student_kappa, student_size, student_time]

plt.figure(figsize=(12, 6))
x = np.arange(len(metrics))
width = 0.35

plt.bar(x - width/2, teacher_values, width, label='EfficientNet (Teacher)')
plt.bar(x + width/2, student_values, width, label='MobileNetV2 (Student)')

plt.xlabel('Metrics')
plt.title('Teacher vs Student Model Comparison')
plt.xticks(x, metrics)
plt.legend()

plt.tight_layout()
plt.show()
```
